```python
years=[]
for i in range(len(cpi_monthly_copy)):
    years.append(cpi_monthly_copy['year'][i][0:4])

months=[]

for i in range(len(cpi_monthly_copy)):
    months.append(cpi_monthly_copy['year'][i][5:7])

year_month = []

for i in range(len(years)):
    year_month.append(years[i] + '-' + months[i])
    
year_month_date = []

for i in range(len(year_month)):
    year_month_date.append(datetime.datetime.strptime(year_month[i], '%Y-%m').date())

cpi_monthly_copy['year'] = year_month_date
```

```python
def detect_outliers_iqr(data):
    q1 = cpi_monthly_copy[data].quantile(0.25)
    q3 = cpi_monthly_copy[data].quantile(0.75)
    # print(q1, q3)
    IQR = q3-q1
    lwr_bound = q1-(1.5*IQR)
    upr_bound = q3+(1.5*IQR)
    # print(lwr_bound, upr_bound)
    return {data:(lwr_bound,upr_bound)}
```

```python
detect_outliers_iqr('Afghanistan')
```
